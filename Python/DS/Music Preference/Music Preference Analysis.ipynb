```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

music_data = pd.read_csv('MusicPreference.csv') 
music_data
```

```python
#Creating an input dataset that only contains the intended input (age and gender)
input_data = music_data.drop(columns = ['Genre'])
input_data
```

```python
#Creating the output dataset that contains the output column (genre)
output = music_data['Genre']
output
```

```python
#building the model with the DTC algorithm

model = DecisionTreeClassifier()
model.fit(input_data, output)

predictions = model.predict([ [31, 1], [33, 1] ])
predictions
```
